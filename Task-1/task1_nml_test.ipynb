```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use("seaborn-whitegrid")
df = pd.read_csv('bank_marketing_test.csv')
df.head()
```

```python
df.iloc[:,0] = df.iloc[:,0].astype('float64')
```

```python
X = df.copy()
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()
discrete_features = X.dtypes == 'int64'
print(discrete_features)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
X['campaign'] = imputer.fit_transform(X[['campaign']])
```

```python
from sklearn.preprocessing import MinMaxScaler

X = X.drop(['loan'], axis=1)
X = X.drop(['day_of_week'], axis=1)
scaler = MinMaxScaler()
X.iloc[:, -5:] = scaler.fit_transform(X.iloc[:, -5:])
X.head()
```

```python
X.to_csv('normalized_test.csv', index=False)
```
